````python
df_last["분양가격"] =  pd.to_numeric(df_last["분양가격(㎡)"],errors='coerce') # 소수로된 세로운 컬럼생성 
````

````python
df_last["분양가격"].max()
````

````python
df_last["평당분양가격"]=df_last["분양가격"] * 3.3 # df_first와 단위를 맞춰줌 
````

````python
df_last["규모구분"].unique()
````

````python
df_last["전용면적"]=df_last["규모구분"].str.replace("전용면적","") # 꼭 str. 쓰고 replace해야함
df_last["전용면적"]=df_last["전용면적"].str.replace("초과","~")
df_last["전용면적"]=df_last["전용면적"].str.replace("이하","")
df_last["전용면적"]=df_last["전용면적"].str.replace(" ","").str.strip() # str.strip() 앞뒤의 공백도 제거 
df_last["전용면적"]
````
